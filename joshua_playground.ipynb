```python
def get_nearby_industrial_sites(lat, lon, radius=100):
    overpass_url = "http://overpass-api.de/api/interpreter"
    overpass_query = f"""
    [out:json];
    (
      node["landuse"="industrial"]({lon - radius},{lat - radius},{lon + radius},{lat + radius});
      way["landuse"="industrial"]({lon - radius},{lat - radius},{lon + radius},{lat + radius});
      relation["landuse"="industrial"]({lon - radius},{lat - radius},{lon + radius},{lat + radius});

    );
    out center;
    """
    response = requests.get(overpass_url, params={'data': overpass_query})
    # data = response.json()
    return response

industrial_sites = get_nearby_industrial_sites(31.52875, 74.330625)
industrial_sites
```

```python
import requests

def get_nearby_data(lat, lon, radius=10000, purpose='"landuse"="industrial"'):
        overpass_url = "http://overpass-api.de/api/interpreter"
        query = f"""
        [out:json];
        (
        node[{purpose}](around:{radius},{lat},{lon});
        way[{purpose}](around:{radius},{lat},{lon});
        relation[{purpose}](around:{radius},{lat},{lon});
        );
        out center;
        """
        response = requests.get(overpass_url, params={'data': query})
        data = response.json()
        return data

# Get nearby cities around the given coordinates 
cities = get_nearby_data(33.51333333333333, -87.20333333333333, 10000, purpose='"landuse"="landfill"')
ind_en_list = [item['tags']['name:en'] for item in cities['elements'] if 'name:en' in item['tags']]
# [item for item in cities['elements']]
cities
```
